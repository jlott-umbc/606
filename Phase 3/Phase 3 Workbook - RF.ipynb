# Random Forest Classifier

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state = 1)
```

```python
rf = RandomForestClassifier(n_estimators= 200)

rf = rf.fit(x_train, y_train)

y_pred = rf.predict(x_test)
```

```python
print(metrics.accuracy_score(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

The random forest classifier seemed to work much better if I did not remove any attributes.  The more I removed attributes, the lower the performance was.  By adjusting the number of estimators I was able to improve performance flightly from around 67% to around 71%.
